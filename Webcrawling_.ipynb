```python
# https://www.melon.com/chart/index.htm
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.melon.com/chart/index.htm"

# 헤더 설정
headers = {
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
print(response.status_code)
```

```python
soup = bs(response.text, 'html.parser')
#response.text 웹사이트내 html 태그들
#뷰티풀스프파싱

songs = soup.find_all('tr', class_='lst50') #lst50이름으로 클래스 속성을 가지는 tr태그들의 집합
print(songs)
```

```python
#곡과 가수이름 찾기
song_list = []
for tr in songs:
  rank_number = tr.find_all('span', class_='rank')[0].text
  rank_text = tr.find_all('div', cass_='wrap_song_info')[0]#find all 은 태그를 여러개 ??는 1개여도 ???
  #print(rank_text)
  #print("==") html정보 뜸
  title = rank_text.find_all('div',class_='rank01')[0].text.strip()
  singer = rank_text.find_all('div',class_='rank02')[0].find('a').text.strip()
  print(f"{rank_number} : {title} - {singer}")
  song_list.append([rank_number,title,  singer])
  #
```

```python
import csv
f = open('/content/drive/mydrive/python_summer/melon.csv','w', newline="", encoding="utf-8") #파일 하나 만든 것?
#이 경로의 파일에 접근,utf-8인코딩 방식으로 파일 내에 요소를 쓰겠다
writer = csv.writer(f)
col_name = ['순위', '노래제목', '가수명'] #테이블 콜롬명을 설정
writer.writerow(col_name)  #행 하나 추가
writer.writerows(song_list)
f.close()
```
