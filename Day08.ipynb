```python
# Numpy, Pandas, Matplotlib을 이용해서 데이터 시각화
# 필요한 library
# Numpy : 수치 데이터를 조작
# Pandas : 데이터 분석
# Matplotlib : 그래프
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 랜덤 생성기 초기화
np.random.seed(0)

data_size = 100
x = np.linspace(0, 10, data_size)
y = np.sin(x) + np.random.normal(scale=0.5, size=data_size) # 고르게 분포된 x값에 대해 랜덤한 y값을 출력

df = pd.DataFrame({
    'X':x,
    'Y':y
})
print(df)
```

```python
plt.style.use('seaborn') # 그래프 스타일 설정
plt.figure(figsize=(10, 6)) # 그래프의 크기 설정
plt.plot(df['X'], df['Y'], label="Random Pos Graph", color="b", marker='o') # 꺾은선그래프 생성

plt.title('Random Pos Graph')   # 그래프 제목
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.legend()
plt.show()  # 설정된 그래프 출력
```
